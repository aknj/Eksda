## 3.1. Histogramy, wykresy pudełkowe i punktowe

```python
# macierz danych o 1000 obiektach o trzech atrybutach ilościowych o losowych wartościach
df6 = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000),
                    'c': np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
# generowanie wykresu, parametr 'alpha' to przenikanie słupków, 'bins' to liczba słupków (standardowo bins = 10) 
#    'title' definiuje tytuł wykresu
df6.plot.hist(alpha = 0.5, bins = 15, title = 'histogram')
```

```python
# histogram skumulowany, pojedynczy atrybut, orientacja pozioma
df6['a'].plot.hist(orientation='horizontal', cumulative=True)
```

```python
# histogramy atrybutów na osobnych wykresach
df6.hist(color='b', alpha=0.5, bins=50)
```

```python
# Wykres pudełkowy poszczególnych atrybutów
df6.plot.box()
```

```python
# Wykres punktowy pary atrybutów atrybutów
df6.plot.scatter(x='a',y='b')
```

```python
# Wykres punktowy dwóch atrybutów, wielkość punktu proporcjonalna do trzeciego
df6.plot.scatter(x='a', y='b', s=df6['c']*20+80);
# Dlaczego w powyższym wywołaniu jest mnożenie przez 20 i dodawanie 80 ?
```

```python
# Inna forma prezentacji tych samych danych - tym razem trzeci atrybut wpływa na kolor komórki
df6.plot.hexbin(x='a', y='b', C='c', gridsize=25)
```

## 3.2. Inne rodzaje wykresów

```python
# szereg czasowy zmian dziennych
zd = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2014', periods=1000))
# szereg wartości aktualnych parametru w kolejnych dniach
# cumsum to suma skumulowana wartości od początku do danego dnia
wa = zd.cumsum()
wa.plot()
```

```python
# taka sama sytuacja, ale obserwacja czterech atrybutów zmiennych w czasie
zd = pd.DataFrame(np.random.randn(1000, 4), index=zd.index, columns=list('ABCD'))
wa = zd.cumsum()
wa.plot();
```
